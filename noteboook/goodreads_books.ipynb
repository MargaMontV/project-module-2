```python
books_2['ISBN 13'] = books_2['isbn13'].astype(str)
books_2['ISBN 13'] = books_2['ISBN 13'].str.split(".", 
    expand = True)[0]
books_2['Author'] = books_2['authors'].str.split("/", 
    expand = True)[0]
books_2['Nº of pages'] = books_2['  num_pages'].astype(int)
books_2['Nº of ratings'] = books_2['ratings_count'].astype(int)
books_2['Rating'] = books_2['average_rating'].astype(str)
books_2['Rating'] = books_2['Rating'].str.replace(".", ",")
books_2['Nº of reviews'] = books_2['text_reviews_count'].astype(int)
books_2['Publication date'] = pd.to_datetime(books_2['publication_date'])
books_2.head(3)
```

```python
books_2.info()
```

```python
books_3 = books_2[['ISBN 13', 'title', 'Author', 'Nº of pages', 'Publication date', 
                   'Rating', 'Nº of ratings', 'Nº of reviews']]
books_3.head(3)
```

```python
books_3.columns = ['ISBN 13', 'Book title', 'Author', 'Nº of pages', 
    'Publication date', 'Rating', 'Nº of ratings', 'Nº of reviews']
books_3.head()
```
